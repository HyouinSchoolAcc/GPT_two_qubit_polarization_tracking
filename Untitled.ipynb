```python
import pandas as pd
data_dir = "C:/Users/yueze/Desktop/GPT_two_qubit_polarization_tracking/20240610_2056.csv"
df = pd.read_csv(data_dir)
print (df.head())
```

```python
import matplotlib.pyplot as plt
sampled_df = df.iloc[::100000, :]

# Plot each column in a separate graph
fig, axs = plt.subplots(6, 1, figsize=(10, 20))

for i, col in enumerate(sampled_df.columns):
    axs[i].plot(sampled_df.index, sampled_df[col], marker='o')
    axs[i].set_title(f'{col} values every 100,000th element')
    axs[i].set_xlabel('Index')
    axs[i].set_ylabel(col)

plt.tight_layout()
plt.show()
```
